# Assignmnt : Create a dataframe with 5 rows and 4 columns randomly and draw different graphs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cols = ['Col'+str(c) for c in range(1,5)]
cols
```

```python
d = pd.DataFrame(np.random.random((5,4)),columns = cols)
d
```

```python
d.plot(linestyle='dashed',title='Random data Plotting')
```

```python
d.plot(kind="area",title="Area chart for random data")
```

```python
d.plot(kind='pie',subplots=True,autopct='%.f%%',layout=(2,2),title='Random data chart',figsize=(10,10))
plt.legend()
plt.show()
```

```python
d.T.plot(kind='pie',subplots=True,autopct='%.f%%',layout=(3,2),title='Transposed random data chart',figsize=(10,10))
plt.legend()
plt.show()
```
